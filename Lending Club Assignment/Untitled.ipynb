```python
import pandas as pd
import numpy as np

po = pd.read_csv("loan.csv", encoding = "ANSI")
po
```

```python
po.info()
po.shape
```

```python
po = po.dropna(axis=1, how='all')
po.shape
```

```python
# FILLING NULL VALUES IN REVOL_UTIL WITH 0% SINCE MOST OF REVOL_BAL CORRESPONDING IS 0
po.dtypes
po['revol_util'].tail(20)
po[po['revol_util'].isnull()]
po.loc[po['revol_util'].isnull(), ['revol_util']] = "0%"
po['revol_util'].isnull().sum()
```

```python
#Converting int_rate column from string to float
po.int_rate = po.int_rate.apply(lambda x: x.replace('%',''))
po.int_rate = pd.to_numeric(po.int_rate)
```

```python
po.revol_util= po.revol_util.apply(lambda x: x.replace('%',''))
po.revol_util = pd.to_numeric(po.revol_util)
```

```python
po.revol_util
```

```python
po.loc[po.emp_length.isnull(),['emp_title','emp_length','annual_inc']]
po.emp_length = po.emp_length.apply(lambda x: x.replace(" years","") if x.notnull() else x)
```

```python
pochargedoff = po[po.loan_status == "Charged Off"]
pogood = po.loc[(po.loan_status == "Fully Paid") | (po.loan_status == "Current"), :]
pocurrent = po[po.loan_status == "Current"]
pogood
```
