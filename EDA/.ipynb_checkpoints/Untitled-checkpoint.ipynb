```python
def is_positive(x):
    if x == "S":
        return 0
    else:
        return x

jio = pd.read_csv("nas.csv")
jio['Siblings'] = jio['Siblings'].apply(lambda x: x[0])
jio['Siblings'] = jio['Siblings'].apply(is_positive)
jio['Siblings'] = pd.to_numeric(jio['Siblings'])
```

```python
tot = jio.pivot_table(values = 'Siblings', 
                      index = 'Mother.edu',
                      aggfunc = lambda x: x.mode().iat[0])
tot
jio.columns
```

```python
jio['Science..'] = pd.to_numeric(jio['Science..'])
nas_data_sc = jio[jio['Science..'].notnull()]
ag = nas_data_sc.pivot_table(values = 'Science..',
                            columns = 'Age',
                            index = 'Father.edu',
                            aggfunc = 'mean')
ag
```

```python
ag.describe()
```

```python
gr = pd.read_csv("grades.csv")
gr.head(10)
def iszip(x):
    y= x.split(".")[3]
    if y == "zip":
        return 1
    else:
        return 0
    
gr['zip_file'] = gr['submission'].apply(iszip)
gr
```

```python
gr.zip_file.sum()/len(gr.index)
```

```python
gr.dtypes
```

```python
gr['submit_date'] = gr['submit_time'].apply(lambda x:x.split('-')[0])
gr.submit_date = pd.to_datetime(gr.submit_date)
gr.dtypes
gr
```

```python
import datetime
d1 = datetime.datetime(2017, 1, 3)
gr['submit_on_time'] = gr['submit_date'].apply(lambda x : x <= d1)
len(gr.index) - gr.submit_on_time.sum()

```

```python
mst = gr.pivot_table(values = 'submit_on_time',
                            index = 'submit_date',
                            aggfunc = 'sum')
mst
gr
```

```python
 
def dates(x):
    y = x.split('-')[1]
    m1 = y.split(':')[0]
    fin = m1+(":")+("00")
    return fin


gr['submit_hour'] = gr['submit_time'].apply(dates)
# gr.submit_hour = pd.to_datetime(gr.submit_hour)
gr.dtypes
gr
gr['pos'] = 1
gr
```

```python
mst1 = gr.pivot_table(values = 'pos',
                            index = 'submit_hour',
                            aggfunc = 'sum')
mst1
```
